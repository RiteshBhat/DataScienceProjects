```python
import pandas
from pandas import DataFrame
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plot
```

```python
data = pandas.read_csv('Movie Data.csv')
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plot.figure(figsize=(10,6))
plot.scatter(X, y)
plot.title('Film Cost vs Revenue')
plot.xlabel('Budget')
plot.ylabel('Revenue')
plot.ylim(0,3000000000)
plot.xlim(0,450000000)
plot.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plot.figure(figsize=(10,6))
plot.scatter(X, y, alpha=regression.coef_/10)
plot.plot(X, regression.predict(X), color='red', linewidth=5)
plot.title('Film Cost vs Revenue')
plot.xlabel('Budget')
plot.ylabel('Revenue')
plot.ylim(0,3000000000)
plot.xlim(0,450000000)
plot.show()
```

```python
regression.score(X, y)
```
